In matrix notation, the analytic solution to the linear regression problem, which we solved with gradient descent in the lecture, is given by: $$\hat{\gamma}=(X'X)^{-1}X'y, \quad \text{where} \quad X=\begin{pmatrix} 1 & x_{1} \\ 1 & x_{2} \\ \vdots & \vdots \\ 1 & x_{n} \end{pmatrix} \quad \text{and} \quad \hat{\gamma} = \begin{pmatrix}\hat{\alpha} \\ \hat{\beta}\end{pmatrix}$$

This formula allows us to study two concepts that we have not discussed in the lecture so far. The __transpose__ of a matrix $A$, denoted $A'$ or $A^T$, is obtained by reflecting $A$ about its main diagonal (the diagonal from the upper left to the lower right). In other words, the entry in the $i$th row and $j$th column of $A$ becomes the entry in the $j$th row and $i$th column of $A^T$:

$\left(A^T\right)_{ij} = a_{ji}, \quad \text{for } 1 \leq i \leq n \text{ and } 1 \leq j \leq m.$

To understand the __inverse__ $A^{-1}$ of a matrix intuitively, recall that in the lecture we visualised multiplication by a matrix as a linear transformation that moves points in space around — multiplying by a matrix could stretch, rotate, or flip a space. The inverse of a matrix is simply the transformation that undoes whatever the original matrix did, i.e. brings the points back to their original location. However, not every matrix has an inverse: the usual inverse exists only for square matrices whose transformation can be undone (example in class).

Formally, for a matrix $A$, its inverse $A^{-1}$ (when it exists) is the unique matrix such that multiplying $A$ by $A^{-1}$ yields the identity matrix $I$. In symbols:

$$A A^{-1} = A^{-1} A = I_n, \quad \text{where} \quad I_n = \begin{pmatrix} 1 & 0 & \cdots & 0 \\ 0 & 1 & \cdots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \cdots & 1 \end{pmatrix}$$

Thus, this generalises the scalar relation $a \times \frac{1}{a} = 1$ to matrices.
